**3.Vectors**

Vectors are 1-dimensional array of numbers or terms.Vectors in geometry express the magnitude and direction of a potential change. Vector [5, -6] is to go right 5 and down 6. A vector with more than one dimension is called a matrix
Traditionally, vectors are written as either column vectors






**Vector Representation**



image.png

```
import numpy as np
 
y = np.array([1,2,3])
x = np.array([2.,3.,4.])
 
print(y + x)
print(y - x)
print(y / x)

# results
# y + x = [3, 5, 7]
# y - x = [-1, -1, -1]
# y / x = [.5, .67, .75]
```

**4.Matrices**

 A matrix is a 2-D array of numbers, so each element is identiﬁedby two indices instead of just one.

**Matrix Dimensions**

We define the dimensions of a matrix by rows and columns.

image.png

```
import numpy as np
 
a = np.array([[1, 2, 3],
              [4, 5, 6]])
a.shape == (2, 3)
# a = [[1 2 3]
#     [4 5 6]]
 
b = np.array([[1, 2, 3]])
b.shape == (1, 3)
# b [[1 2 3]]
```

**Matrix Scalar Operations**


Runs scaling operations with matrices as with vectors. Apply scalars to each element in the matrix - add, subtract, divide, multiply.

image.png

```
import numpy as np
 
a = np.array(
[[1, 2],
[3, 4]])
 
print(a + 1)
# [[2,3],
# [4,5]]
```

**Matrix Elementwise Operations**

In order for two matrices to add, subtract or divide, they must have equal dimensions. We combine the relevant values elementwise to generate a new matrix.


image.png

```
import numpy as np
 
a = np.array([
    [1, 2],
    [3, 4]
])
b = np.array([
    [1, 2],
    [3, 4]
])
 
print(a + b)
# [[2, 4],
# [6, 8]]
```

**Numpy Broadcasting**

In Numpy, size requirements for elementwise operations are relieved with a mechanism called broadcasting. The two matrices are compatible if the dimensions (rows and rows, columns and columns) in each matrix meet the following requirements:

* If the dimensions are equal or
* If someone's size is 1-dimensional





```
import numpy as np
 
a = np.array([
    [1],
    [2]
])
b = np.array([
    [3, 4],
    [5, 6]
])
c = np.array([
    [1, 2]
])
# The same number of rows
# Of different columns
# but the size of a is a column so it works
print(a * b)
# [[ 3, 4],
# [10, 12]]
 
# Of the same columns
# Number of different lines
# but c one line so it works
print(b * c)
# [[ 3, 8],
# [5, 12]]
 
# Number of different lines
# Of different columns
# but it works one-dimensional in a or c
print(a + c)
# [[2, 3],
# [3, 4]]
```

**Matrix Hadamard Multiplication**


The Hadamard matrices multiplier is an elementwise operation. A new matrix is created by multiplying the corresponding values positionally.


image.png

```
import numpy as np
 
a = np.array(
    [[2, 3],
     [2, 3]])
b = np.array(
    [[3, 4],
     [5, 6]])
 
# Uses Python's multiplication
print(a * b)
# [[ 6, 12],
# [10, 18]]
```

In numpy, the Hadamard Multiplication can be taken as long as the dimensions of a matrix and a vector meet the broadcasting requirements.

**Matrix Transpose**

> In artificial neural networks, weights and sizes of inputs often do not meet the requirements of matrix multiplication. Matrix transposition provides the method of "rotate" one of the matrices; thus the process adapts to multiplication requirements. The process of transposing a matrix is two steps:

* Rotate the matrix 90 ° to the right
* Reverse the order of items in each row (eg [a b c] [c b a] would)



image.png

```
import numpy as np
 
a = np.array([
[1, 2],
[3, 4]])
 
print(a.T)
# [[1, 3],
# [2, 4]]
```

**Matrix Multiplication**

Matrix multiplication, there are several rules for generating a new matrix to multiply matrices. Not all matrices are suitable for multiplication. In addition, there are some requirements on the dimensions of the resulting matrix output.

* The number of columns in matrix 1 must be equal to the number of rows in matrix 2.

* The product of an M x N matrix and an N x K matrix is an M x K matrix. The new matrix takes the rows of matrix 1 and the dimensions of the columns of matrix 2.

Matrix multiplication relies on dot product to multiply various combinations of rows and columns

**Matrix Multiplication with Numpy**

Numpy uses the function np.dot (A, B) for both vector and matrix multiplication.

```
import numpy as np
 
a = np.array([
[1, 2]
])
# a.shape == (1,2)
b = np.array([
[3, 4],
[5, 6]
])
# b.shape == (2,2)
# Multiplication
mm = np.dot(a, b)
print(mm)
# mm == [13, 16]
# mm.shape == (1,2)
```
